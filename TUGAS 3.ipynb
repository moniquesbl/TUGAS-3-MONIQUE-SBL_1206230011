```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat dataset universitas
university_df = pd.read_csv('Preprocessed World University Rankings 2023 Dataset.csv')
# Memeriksa nama kolom
print("Nama kolom dalam dataset universitas:")
print(university_df.columns)

# Memeriksa tipe data
print("\nTipe data setiap kolom:")
print(university_df.dtypes)

# 1. Bar Chart untuk distribusi skor pendapatan industri
# Kita asumsikan ada kolom 'Industry Income Score'
if 'Industry Income Score' in university_df.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(university_df['Industry Income Score'], bins=30, kde=True, color='purple')
    plt.title('Distribusi Skor Pendapatan Industri')
    plt.xlabel('Industry Income Score')
    plt.ylabel('Frekuensi')
    plt.grid(True)
    plt.show()
else:
    print("Kolom 'Industry Income Score' tidak ditemukan dalam dataset universitas.")

# 2. Box Plot untuk menunjukkan hubungan antara 'Teaching Score' dan 'Location' (misalkan ada kolom tersebut)
if 'Location' in university_df.columns and 'Teaching Score' in university_df.columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=university_df, x='Location', y='Teaching Score', palette='Set3')
    plt.title('Distribusi Skor Pengajaran Berdasarkan Lokasi')
    plt.xlabel('Location')
    plt.ylabel('Teaching Score')
    plt.xticks(rotation=90)
    plt.show()
else:
    print("Kolom 'Location' atau 'Teaching Score' tidak ditemukan dalam dataset universitas.")

# 3. Scatter Plot untuk hubungan antara 'Research Score' dan 'Overall Score'
if 'Research Score' in university_df.columns and 'Overall Score' in university_df.columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=university_df, x='Research Score', y='Overall Score', hue='Location', palette='viridis')
    plt.title('Hubungan antara Research Score dan Overall Score')
    plt.xlabel('Research Score')
    plt.ylabel('Overall Score')
    plt.grid(True)
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.show()
else:
    print("Kolom 'Research Score' atau 'Overall Score' tidak ditemukan dalam dataset universitas.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat dataset universitas
university_df = pd.read_csv('Preprocessed World University Rankings 2023 Dataset.csv')
# 1. Histogram untuk distribusi Teaching Score
plt.figure(figsize=(10, 6))
sns.histplot(university_df['Teaching Score'], bins=30, kde=True, color='blue')
plt.title('Distribusi Skor Mengajar')
plt.xlabel('Skor Mengajar')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()

# 2. Box Plot untuk perbandingan Teaching Score berdasarkan Rasio Jenis Kelamin
plt.figure(figsize=(12, 6))
sns.boxplot(data=university_df[['Teaching Score', 'Female Ratio', 'Male Ratio']], palette='Set2')
plt.title('Distribusi Skor Mengajar Berdasarkan Rasio Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Skor Mengajar')
plt.xticks(ticks=[0, 1, 2], labels=['Skor Mengajar', 'Rasio Perempuan', 'Rasio Laki-laki'])  # Menyesuaikan label
plt.show()

# 3. Scatter Plot untuk hubungan antara Research Score dan Industry Income Score
plt.figure(figsize=(10, 6))
sns.scatterplot(data=university_df, x='Research Score', y='Industry Income Score', color='green')
plt.title('Hubungan antara Research Score dan Industry Income Score')
plt.xlabel('Research Score')
plt.ylabel('Industry Income Score')
plt.grid(True)
plt.show()

# 4. Bar Plot untuk distribusi Mahasiswa Internasional
plt.figure(figsize=(10, 6))
sns.histplot(university_df['International Student'], bins=30, kde=True, color='orange')
plt.title('Distribusi Mahasiswa Internasional')
plt.xlabel('Rasio Mahasiswa Internasional')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()

# 5. Pie Chart untuk proporsi Laki-laki dan Perempuan
labels = ['Rasio Perempuan', 'Rasio Laki-laki']
sizes = [university_df['Female Ratio'].mean(), university_df['Male Ratio'].mean()]
colors = ['lightblue', 'lightcoral']
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Proporsi Laki-laki dan Perempuan di Universitas')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Memuat dataset siswa
students_df = pd.read_csv('StudentsPerformance_with_headers.csv')

# 1. Analisis hubungan antara frekuensi membaca dan nilai akhir
# Menyusun data untuk analisis
students_df['Reading frequency'] = students_df['Reading frequency'].astype(float)  # Ubah tipe data jika diperlukan
students_df['GRADE'] = students_df['GRADE'].astype(float)  # Ubah tipe data jika diperlukan

# Menghitung korelasi
correlation_reading_grade, p_value_reading_grade = pearsonr(students_df['Reading frequency'], students_df['GRADE'])

# Menampilkan hasil
print(f"Koefisien Korelasi antara Frekuensi Membaca dan Nilai Akhir: {correlation_reading_grade}")
print(f"P-value: {p_value_reading_grade}")

# Visualisasi hubungan
plt.figure(figsize=(10, 6))
sns.scatterplot(data=students_df, x='Reading frequency', y='GRADE')
plt.title('Hubungan antara Frekuensi Membaca dan Nilai Akhir')
plt.xlabel('Frekuensi Membaca (Buku/Jurnal Non-Ilmiah)')
plt.ylabel('Nilai Akhir')
plt.axhline(y=students_df['GRADE'].mean(), color='r', linestyle='--')
plt.show()

# 2. Analisis hubungan antara Skor Pendapatan Industri dan Lokasi
# Memuat dataset universitas
university_df = pd.read_csv('Preprocessed World University Rankings 2023 Dataset.csv')

# Menampilkan nama kolom dan tipe data untuk pemeriksaan
print("Kolom dalam dataset universitas:")
print(university_df.columns)

# Menghitung rata-rata Skor Pendapatan Industri berdasarkan lokasi
# Kita akan mencoba menggunakan nama kolom lokasi yang sesuai
# Misalkan kita mengelompokkan dengan menggunakan nama kolom yang berisi 'Location'
# kita akan melakukan pemeriksaan lebih lanjut di sini

# Ambil kolom dengan nama yang mengandung "Location"
location_columns = [col for col in university_df.columns if 'Location' in col]
print("Kolom lokasi yang tersedia:", location_columns)

# Menghitung rata-rata Industry Income Score
# Untuk contoh ini, kita bisa menghitung rata-rata berdasarkan salah satu kolom lokasi yang ada
if location_columns:
    location_industry_df = university_df.groupby(location_columns[0])['Industry Income Score'].mean().reset_index()
    print(location_industry_df)

    # Visualisasi rata-rata Skor Pendapatan Industri berdasarkan lokasi
    plt.figure(figsize=(12, 6))
    sns.barplot(data=location_industry_df, x=location_columns[0], y='Industry Income Score', palette='viridis')
    plt.title('Rata-rata Skor Pendapatan Industri berdasarkan Lokasi')
    plt.xlabel(location_columns[0])
    plt.ylabel('Rata-rata Skor Pendapatan Industri')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Tidak ada kolom lokasi yang ditemukan.")
```
